```python
import pandas as pd
```

```python
df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.strip().str.replace(' ','_').str.replace(r'[^\w]','',regex=True).str.lower()
df = df.rename(columns = {'purchase_amount_usd':'purchase_amount'})
```

```python
df.columns
```

```python
# add a column age_group based on exsisitng columns
labels = ['Young Adult', 'Adult', 'Middle-Aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# create column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
    
```

```python
df[['frequency_of_purchases', 'purchase_frequency_days']].head(10)
```

```python
df[['discount_applied', 'promo_code_used']].head(10)
```

```python
df = df.drop('promo_code_used', axis=1)
```
